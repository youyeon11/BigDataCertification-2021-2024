```python
X_train = pd.read_csv(data_path + "1_1_commerce_x_train.csv")
X_test = pd.read_csv(data_path + "1_2_commerce_x_test.csv")
y_train = pd.read_csv(data_path + "1_3_commerce_y_train.csv")
```

```python
X_train = X_train.drop(columns = X_train.select_dtypes(include = 'object').columns, axis =1)
X_test = X_test.drop(columns = X_test.select_dtypes(include = 'object').columns, axis=1)
```

```python
print(X_train.columns)
print("=====================")
print(X_test.columns)
```

예측을 위한 모델 학습 -> 회귀분석 / 의사결정나무 / 랜덤포레스트

```python
# 필요한 라이브러리 불러오기
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
# 회귀 분석을 위한 ID Column 삭제 : pop() 사용 -> 제거와 동시에 반환
X_train_id = X_train.pop("ID")
X_test_id = X_test.pop("ID")
X_train.head()
```

```python
y_train.columns
```

```python
# Split Train & Test Dataset
from sklearn.model_selection import train_test_split
# X_train과 y_train을 학습에 사용 (X_tr: 훈련용, y_tr: 훈련용의 결과, X_val: 검증용, y_val: 검증용의 결과(예측 test로 사용됨))
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train['Reached.on.Time_Y.N'],# 배송 도착 여부에 대한 예측
                                            test_size = 0.2, # Validation Set에 대한 비율
                                            random_state = 1234)
```

로지스틱 회귀분석

```python
# 로지스틱 회귀분석 모델 생성 및 학습
model_lgm = LogisticRegression()
model_lgm.fit(X_tr, y_tr)
```

```python
pred_lgm = model_lgm.predict_proba(X_val) # 예측 결과를 확률로 반환
roc_auc_score(y_val, pred_lgm[ : , 1]) # Class 1에 대해서만 추출 (-> positive class, 배송 도착)
```

```python
pred_lgm
```

의사결정나무

```python
# 의사결정나무 모델 생성 및 학습
model_dt = DecisionTreeClassifier()
model_dt.fit(X_tr, y_tr)
```

```python
pred_dt = model_dt.predict_proba(X_val)
roc_auc_score(y_val, pred_dt[ : , 1])
```

랜덤포레스트

```python
# 랜덤포레스트 모델 생성 및 학습
model_rf = RandomForestClassifier() # 0 ,1
model_rf.fit(X_tr, y_tr)
```

```python
pred_rf = model_rf.predict_proba(X_val)
```

```python
print(pred_rf)
```

```python
roc_auc_score(y_val, pred_rf[ : , 1])
```
